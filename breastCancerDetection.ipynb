```python
df.drop(columns=['Unnamed: 32','id'], inplace=True)
df.T
```

Our target variable `diagnosis` is object and all the feature variables are float64 

```python
df.diagnosis = df.diagnosis == 'M'
df.diagnosis
```

```python
df.diagnosis = df.diagnosis.astype(int)
df.diagnosis  # diagnosis = 'M' => 1,  diagnosis = 'B' => 0
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
mean = ['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave_points_mean','symmetry_mean']
se = ['radius_se','perimeter_se','area_se','compactness_se','concavity_se','concave_points_se']
worst = ['radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave_points_worst','symmetry_worst','fractal_dimension_worst']
```

As we can see that there are outliers in our dataset because of large gap between min and max value of column, applying `np.log1p` on the columns to normalize them.

```python
log = []
```

```python
for col in mean:
    df[col+'_log'] = df[col].apply(np.log1p)
    log.append(col+'_log')
    # display(df[col+'_log'])
```

```python
for col in se:
    df[col+'_log'] = df[col].apply(np.log1p)
    log.append(col+'_log')
```

```python
for col in worst:
    df[col+'_log'] = df[col].apply(np.log1p)
    log.append(col+'_log')
```

```python
from sklearn.model_selection import train_test_split
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
df_full_train = df_full_train.reset_index(drop=True)
df_full_train
```

```python
df_train, df_val = train_test_split(df_full_train, test_size=0.3, random_state=40)
df_train = df_train.reset_index(drop=True)
df_train
```

```python
df_val.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
model = LogisticRegression()
def train(df,y):
    X = df[log].values
    model.fit(X,y)
    return model
```

```python
def predict(df, model):
    X = df[log].values
    y = df.diagnosis.values
    y_pred = model.predict_proba(X)[:,1]
    diagnosis = y_pred >= 0.5
    acc = accuracy_score(y,diagnosis)
    con_mat = confusion_matrix(y, diagnosis)
    return acc, con_mat
```

```python
train(df_train, df_train.diagnosis.values)
```

```python
predict(df_val,model) # accuracy has improved by 3% as we trained on the features of log
```
